```python
import random 

RCT_predicted_df = RCT_Tagger_output_df[(RCT_Tagger_output_df["PREDICTION"] =="RANDOMIZED_CONTROLLED_TRIAL") & 
                                        (RCT_Tagger_output_df["CONFIDENCE"] >=0.95)]

PMID_RCT_predicted = set(RCT_predicted_df["PMID"])
PMID_RCT_predicted = list(PMID_RCT_predicted)

PMID_RCT_str_predicted = []
for PMID in PMID_RCT_predicted:
    PMID = str(PMID)
    PMID_RCT_str_predicted.append(PMID)
    
print ("Number of RCTs from Tagger we are sampling: ", len(PMID_RCT_str_predicted))
```

```python
# Check titles
import pandas as pd
import os
import random
import collections
import urllib
from Bio import Entrez
from xml.dom import minidom
import xml.etree.ElementTree as ET


bad_PMIDs = []

PMID_list = []
Title_list = []
Publication_Types_list = []
count = 0

for PMID in PMID_RCT_str_predicted:
    if PMID != "":
        article_pmid = PMID
        article_title = ""
        pt_lst = []
        
        if count%500 == 0:
            print (count)

        link = "https://eutils.ncbi.nlm.nih.gov/entrez/eutils/efetch.fcgi?db=pubmed&api_key=1a7d18b4a3b452b6f1d688fa1ad4282bfa09&retmode=xml&id=" + str(PMID)

        request = urllib.request.Request(link) 
            
        try:
            result = urllib.request.urlopen(request)
        except:
            bad_PMIDs.append(PMID)
            pass
            
        try:
            tree = ET.parse(result)
        except ET.ParseError as e:
            pass
                        
        doc = tree.getroot()

        #get title
        title = doc.find (".//ArticleTitle")
        if title is None:
            title_sub = doc.find (".//ArticleTitle/i")
            if title_sub is None:
                article_title = "None"
            else:
                article_title = title_sub.text
        else:
            article_title = title.text
            
              
        # get publication type
        publication_type_list = doc.find(".//PublicationTypeList")
        if publication_type_list is not None:      
            publicationtypes = publication_type_list.getchildren()
            for publication_type in publicationtypes:
                pt_lst.append(publication_type.text)
        else:
            pt_lst.append("None")
            
        PMID_list.append(article_pmid)
        Title_list.append(article_title)
        Publication_Types_list.append(pt_lst)
        
        count+=1 
        
all_data_withmetadata_df = pd.DataFrame(list(zip(PMID_list,Title_list, Publication_Types_list)), 
                   columns =['PMID','Title','Publication Types']) 


print ("Number of PMIDs that don't have links: ", len(bad_PMIDs))
print ("Number of PMIDs that have links: ", len(all_data_withmetadata_df))

display(all_data_withmetadata_df.head(10))
```

```python
# Get papers from the list
all_data_df = pd.DataFrame()
count = 0
for PMCID in PMID_RCT_str_predicted:
    if count%500 == 0:
        print (count)
    
    PMCID_df = pd.DataFrame()
    with open(data_path + PMCID + ".json") as f:
        data = json.load(f)  
        for sentence_item in data:            
            sentence_df = pd.DataFrame.from_dict(sentence_item,orient='index')
            sentence_df= sentence_df.transpose()

            PMCID_df = pd.concat([PMCID_df,sentence_df])
            
    PMCID_df["PMID"] = PMCID
    PMCID_df = PMCID_df[["PMID","sentence","sid","span","section","token","lemma","pos"]]
    PMCID_df = PMCID_df.reset_index(drop=True)

    all_data_df = pd.concat([all_data_df,PMCID_df])
    all_data_df = all_data_df.reset_index(drop=True)
    count +=1 
display(all_data_df.head(100)) 
```

```python
snorkel_data_file = "/efs/CONSORT2/data/snorkel_data/all_methodological_data_05212021.csv"

snorkel_data_df = pd.read_csv(snorkel_data_file)
snorkel_data_df = snorkel_data_df[["PMID","sid", "sentence","section"]]
snorkel_data_df.columns = ["PMID","sid", "text","section_hierarchy"]
display(snorkel_data_df.head(5))

print (len(snorkel_data_df))
```

```python
# Preprocessing snorkel data

# Remove rows that are not string
snorkel_data_df = snorkel_data_df[~snorkel_data_df["text"].isnull()]
print (len(snorkel_data_df))

# Remove rows with too short text
snorkel_data_df = snorkel_data_df[~snorkel_data_df['text'].apply(lambda x: len(x.split(' ')) < 5)]
print (len(snorkel_data_df))

# Remove rows that only numbers
snorkel_data_df = snorkel_data_df[~snorkel_data_df['text'].str.isdigit()]
print (len(snorkel_data_df))

snorkel_data_df = snorkel_data_df.reset_index()
snorkel_data_df = snorkel_data_df[["PMID","sid","text","section_hierarchy"]]

# Preprocess convert all to lower case
snorkel_data_df["text"] = snorkel_data_df["text"].str.lower()
snorkel_data_df["section_hierarchy"] = snorkel_data_df["section_hierarchy"].str.lower()

display(snorkel_data_df.head(10))

# Remove sentences that are section header
preprocessed_snorkel_data_df = snorkel_data_df[~(snorkel_data_df["text"].isin(snorkel_data_df["section_hierarchy"]))]

# Remove sentences that are Figure/Table
preprocessed_snorkel_data_df = preprocessed_snorkel_data_df[~(preprocessed_snorkel_data_df["text"].str.contains("figure|table|shown in"))]
print (len(preprocessed_snorkel_data_df))

display(preprocessed_snorkel_data_df.head(100))
```

```python
preprocessed_snorkel_data_df = preprocessed_snorkel_data_df.reset_index()
preprocessed_snorkel_data_df.to_csv("/efs/CONSORT2/data/snorkel_data/all_methodological_data_preprocessed_05232021.csv", index = False)
```
